# Penguins Data Set

```python
#importing packages
import itertools
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#Reading in the dataset.
#This is a famous dataset that comes preloaded as a part of seaborne graphing library

penguins_size = sns.load_dataset("penguins")
#penguins_size = pd.read_csv("penguins_size.csv")
```

```python
penguins_size.columns #looking at each column
```

```python
#looking at data type of each column
penguins_size.dtypes
```

```python
#Seeing unique values in the dataset
cols_unique_vals = ['species', 'island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm',
       'body_mass_g', 'sex'] #making a list of columns to further explore

for col in cols_unique_vals : #looping through each name in the above list
  print(col) #print the name of the column
  print(penguins_size[col].unique()) #prin the unique values in that column
```

```python
#Histogram of Body Mass Using the SeaBorne graphing package (instead of matplotlib)
fig, ax = plt.subplots()

sns.histplot(x=penguins_size['body_mass_g'], color = "black")

ax.set_ylim(bottom=0)
ax.set_xlim(left=0)
plt.xlabel('Body Mass in Grams')

fig.show()
```

#Transforming the data and adding a new column
Let's pretend that we want the body mass of the penguins in pounds instead of grams. The code below adds a new column to the data table that will be the penguins weight in pounds and then the subsequent panel will plot the histogram in pounds.

```python

#Creates a new column called body_mass_lb and generates the value from the body mass in g converted to lb.
penguins_size['body_mass_lb'] = penguins_size['body_mass_g']*.00220462

#Note the new column
penguins_size.head(3)
```

```python
#Histogram of Body Mass in pounds (lb) Using the SeaBorne graphing package (instead of matplotlib)
fig, ax = plt.subplots()

sns.histplot(x=penguins_size['body_mass_lb'])

ax.set_ylim(bottom=0)
ax.set_xlim(left=0)
plt.xlabel('Body Mass in Pounds')

fig.show()
```

#Boxplots by weight of penguins broken down by species

Learn more about boxplots [here](https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/box-plot-review#:~:text=A%20box%20and%20whisker%20plot%E2%80%94also%20called%20a%20box%20plot,the%20box%20at%20the%20median.)

```python
plt.figure()

sns.boxplot(data=penguins_size, x="species", y="body_mass_g")
plt.title('Boxplot of Body Mass of Species of Penguin')
plt.xlabel('Species')
plt.ylabel('Body Mass')
plt.show()
```

```python
#Adding another dimension to the plot is easy.
#Each species body mass is further split by sex

sns.boxplot(data=penguins_size, x="species", y="body_mass_g", hue="sex")
```

#Violin Plots
Violin plots show the same information as the boxplot as well as show the distribution of the data.

Read more about violin plots [here](https://https://en.wikipedia.org/wiki/Violin_plot)

```python
plt.figure()

sns.violinplot(data=penguins_size, x="species", y="body_mass_g")
plt.title('Violin Plot of Body Mass of Species of Penguin')
plt.xlabel('Species')
plt.ylabel('Body Mass')
plt.show()
```

#Bar Plots in Seaborne
Bar plots in seaborne will take a column of data and automatically plot the average and standard deviation from it. Pretty nice.

```python
plt.figure()
sns.barplot(data=penguins_size, x="island", y="body_mass_g")

plt.title('Body Mass of Penguins of Various Islands')
plt.xlabel('Island')
plt.ylabel('Body Mass in Grams')
plt.show()
```

```python
#break down body mass by sex
plt.figure()
sns.barplot(data=penguins_size, x="island", y="body_mass_g", hue="sex")

plt.title('Body Mass of Penguins of Various Islands')
plt.xlabel('Island')
plt.ylabel('Body Mass in Grams')
plt.show()

```
